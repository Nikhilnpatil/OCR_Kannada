```python
import pandas as pd
import numpy as np
import os
import cv2
```

```python
PATH = "../archive/Img"
```

```python
df = pd.read_csv(r"C:\Users\sidha\Desktop\Tarran\Github\itpm\archive\kannada.csv")
```

```python
map = dict(df.values)
map
```

```python
X = []
y = []
for img in os.listdir(PATH):
    X.append(img)
    y.append(map["Img/"+img])
print(len(X),len(y))
```

```python
y = [str(x-1) for x in y]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=420)
```

```python
def shiftImgs(dataX,dataY,type):
    Pth = "C:\\Users\sidha\\Desktop\\Tarran\\Github\\itpm\\data"
    OriginalPath = "C:\\Users\sidha\\Desktop\\Tarran\\Github\\itpm\\archive\\Img"
    for index in range(0,len(dataX)):
        if os.path.exists(os.path.join(Pth,type,dataY[index])) == False:
            os.mkdir(os.path.join(Pth,type,dataY[index]))
        img = cv2.imread(os.path.join(OriginalPath,dataX[index]))
        img_scaled = cv2.resize(img,(128,128))
        img_grayscale = cv2.cvtColor(img_scaled, cv2.COLOR_BGR2GRAY)
        cv2.imwrite(os.path.join(Pth,type,dataY[index],dataX[index]),img_grayscale)
```

```python
shiftImgs(X_train,y_train,"train")
```

```python
shiftImgs(X_test,y_test,"val")
```
